```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('./inventory sales.csv')
data1
```

```python
# Assuming 'data' is your DataFrame containing the sales data

# Filter the DataFrame
data2 = data1[data1['product_title'].str.contains('polo|tee', case=False)]

# Display the filtered DataFrame
data1
```

```python
data4 = data1[['product_variant_title', 'quantity_sold']]
data4
```

```python
# Assuming 'data4' is your DataFrame

# Replace values
data4['product_variant_title'] = data4['product_variant_title'].replace({'X-Large': 'XL', 'XX-Large': 'XXL', 'Medium': 'M', 'Large': 'L', 'Small': 'S'})

# Delete rows where 'Default Title' appears in 'Size' column
data4 = data4[data4['product_variant_title'] != 'Default Title']

# Display the modified DataFrame
data4
```

```python
# Assuming 'data4' is your DataFrame

# Group by 'product_variant_title' and sum the quantities
data4 = data4.groupby('product_variant_title', as_index=False)['quantity_sold'].sum()

# Display the aggregated data
data4
```

```python
# Preprocessing - Encoding categorical variables (e.g., shirt sizes) if needed
data4 = pd.get_dummies(data4, columns=['product_variant_title'])
```

```python
# Splitting data into features (X) and target variable (y)
X = data4.drop('quantity_sold', axis=1)  # Features
y = data4['quantity_sold']  # Target variable
```

```python
# Creating and training the linear regression model
model = LinearRegression()
model.fit(X,y)
```

```python
restock_quantity = 10
predicted_quantities = model.predict(X)  # Predicting quantities based on the entire dataset
sizes = X.columns
restock_sizes = []
restock_quantities = []

# Calculate how many shirts of each size to restock
for size in sizes:
    predicted_quantity_for_size = sum(predicted_quantities[X[size] == 1])
    restock_size_quantity = max(0, restock_quantity * predicted_quantity_for_size / sum(predicted_quantities))
    restock_sizes.append(size)
    restock_quantities.append(round(restock_size_quantity))

# Display the restocking plan
restocking_plan = pd.DataFrame({'Size': restock_sizes, 'Quantity to Restock': restock_quantities})
restocking_plan
```
